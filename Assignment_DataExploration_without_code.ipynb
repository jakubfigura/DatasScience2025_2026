```python
df = pd.read_csv('kc_house_data.csv')
df.head(10)
```

```python
categorical_values = ['id', 'date', 'waterfront', 'zipcode','view', 'condition', 'grade']
house_categorical_df = df[categorical_values]
house_numerical_df = df.drop(categorical_values, axis = 1)
house_categorical_df.head()

```

## Exercise 2
Using scatter plot:  
a) Show the relation of square footage and price.  
b) Check how price is influenced by apartment's grade. Find other variables correlated with price.  
c) Check how good apartments are distributed over the city. First, plot zipcode versus price. Then, use apartments coordinates (longitude and lattitude) in order to show where apartments of good grade are located. 

```python
price = house_numerical_df['price']
square_footage = house_numerical_df['sqft_living']
plt.scatter(square_footage, price, color = 'green', alpha= 0.4)
plt.title('Square footage ~ Price')
plt.xlabel('Square footage')
plt.ylabel('Price')
plt.show()
```

```python
for col in house_numerical_df.columns:
    if col != 'price': 
        plt.figure(figsize=(6,4))
        sns.scatterplot(x=house_numerical_df[col], y=house_numerical_df['price'])
        plt.title(f'Price vs {col}')
        plt.show()
```

```python
corr_matrix = house_numerical_df.corr()
print(corr_matrix['price'])
```

```python
price = house_numerical_df['price']
grades = house_categorical_df['grade']
plt.scatter(grades, price, color = 'tomato', alpha= 0.4)
plt.title('Grade ~ Price')
plt.xlabel('Grade')
plt.ylabel('Price')
plt.show()

```

```python
price = house_numerical_df['price']
zip_code = house_categorical_df['zipcode']
plt.scatter(zip_code, price, color = 'magenta', alpha= 0.4)
plt.title('Zip Code ~ Price')
plt.xlabel('Zip Code')
plt.ylabel('Price')
plt.show()

```

```python
long = house_numerical_df['long']
lattitude = house_numerical_df['lat']
plt.figure(figsize=(10,6))
plt.scatter(long, lattitude, cmap='viridis', c=grades, s = 12)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title("Longitude ~ Latitude ")
plt.plot()

```

```python
unique_grades = sorted(house_categorical_df['grade'].unique())
price_grades = df[["price", "grade"]]

data = [price_grades[price_grades['grade'] == g]["price"] for g in unique_grades]

plt.figure(figsize=(10,8))
plt.boxplot(data, tick_labels=unique_grades)
plt.xlabel("Grade")
plt.ylabel("Price")
plt.title("Price Distribution by House Grade")
plt.show()
```

```python
from statsmodels.graphics.mosaicplot import mosaic

def my_mosaic(*args,**kwargs):
    mosaic(kwargs['data'], list(args),
           labelizer=lambda k: '',
           ax=plt.gca())

g = sns.FacetGrid(df, col='view', col_wrap=3)
g = g.map_dataframe(my_mosaic, 'condition', 'waterfront')
plt.show()


```
